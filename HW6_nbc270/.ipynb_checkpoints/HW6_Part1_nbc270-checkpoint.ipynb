```python
mn_data = mn_data[[u'ZipCode',u'LandUse']]
mn_data.head()
```

```python
bk_data = bk_data[[u'ZipCode',u'LandUse']]
bk_data.head()
```

```python
bkgrouped = bk_data.groupby([u'LandUse']).agg('count')
bkgrouped = bkgrouped.reset_index()
print(bkgrouped)
```

```python
mngrouped = mn_data.groupby([u'LandUse']).agg('count')
mngrouped = mngrouped.reset_index()
print(mngrouped)
```

```python
pd.to_numeric(mngrouped[u'LandUse'], errors='coerce')
mngrouped.head()
```

```python
pd.to_numeric(bkgrouped[u'LandUse'], errors='coerce')
bkgrouped.head()
```

```python
mn_bk = pd.merge(bkgrouped, mngrouped, on = u'LandUse')
mn_bk.head()
```

```python
# mn_bk.columns = {'LandUse':u'LandUse','Brooklyn':u'ZipCode_x','Manhattan':u'ZipCode_y'}
mn_bk.columns = ['LandUse', 'Brooklyn', 'Manhattan']
mn_bk.head()
```

```python
code_to_verbose = {1.0: 'One &Two Family Buildings',2.0: 'Multi-Family Walk-Up Buildings',3.0: 'Multi-Family Elevator Buildings',4.0: 'Mixed Residential & Commercial Buildings',5.0: 'Commercial & Office Buildings',6.0: 'Industrial & Manufacturing',7.0: 'Transportation & Utility',8.0: 'Public Facilities & Institutions',9.0: 'Open Space & Outdoor Recreation',10.0: 'Parking Facilities',11.0: 'Vacant Land'}
mn_bk[u'LandUse'] = mn_bk[u'LandUse'].map(code_to_verbose)
mn_bk.head(11)
```

```python
mn_bk['luration'] = mn_bk['ZipCode_y'] / mn_bk['ZipCode_x']
mn_bk.head(11)
```

```python
mn_bk['Manhattan Wins'] = mn_bk['luration'] > 1
mn_bk.head(11)
```
