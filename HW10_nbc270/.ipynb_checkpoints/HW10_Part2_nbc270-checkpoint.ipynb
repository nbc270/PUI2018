# Fit the Data with a Line in Log Space:

```python
#Calculate log of energy, total units, and the square of log of total units
bblnrgdata_cut['lognrg'] = np.log10(bblnrgdata_cut['nrg'])
bblnrgdata_cut['logUT'] = np.log10(bblnrgdata_cut['UnitsTotal'])
bblnrgdata_cut['logUT2']=bblnrgdata_cut['logUT']**2
```

```python
#Replace -Inf, if there are any, with NaN and dropna. This is because the model won't work with Inf values
bblnrgdata_cut = bblnrgdata_cut.replace(-np.Inf, np.nan)
bblnrgdata_cut = bblnrgdata_cut.dropna()
```

```python
#Create the first model (correct model)
OLS1=smf.ols(formula='lognrg~logUT',data=bblnrgdata_cut).fit()
OLS1.summary()
```

```python
#Plot first model over data
pl.figure(figsize=(10,10))
pl.plot(bblnrgdata_cut['logUT'],bblnrgdata_cut['lognrg'],'go')
pl.plot(bblnrgdata_cut['logUT'], OLS1.predict(bblnrgdata_cut),'-', color = 'orange')
pl.xlabel('Independent')
pl.ylabel('Depedent')
pl.title('Figure 9: Anscomby 1 - Linear fit')
```

Figure nine shows energy vs total units per lot in log scale overlayed with a OLS linear fit. The fit appears to run through the mean of the dependent (being energy). The fit does not appear to be very good and should be compared to a transpose of this data.

```python
#Create second model (incorrect one)
OLS2=smf.ols(formula='logUT~lognrg',data=bblnrgdata_cut).fit()
OLS2.summary()
```

```python
#Plot second model over data
pl.figure(figsize=(10,10))
pl.plot(bblnrgdata_cut['lognrg'],bblnrgdata_cut['logUT'],'go')
pl.plot(bblnrgdata_cut['lognrg'], OLS2.predict(bblnrgdata_cut),'-', color = 'orange')
pl.xlabel('Independent')
pl.ylabel('Depedent')
pl.title('Figure 10: Anscomby 2 - Linear fit')
```

Figure ten is the transpose of the data in figure nine with another OLS linear fit. The fit does not sppear to describe the data at all, suggesting that the previous fit was better. 

```python
#Create third model which is a quadratic
OLS3=smf.ols(formula='lognrg~logUT2 + logUT',data=bblnrgdata_cut).fit()
OLS3.summary()
```

```python
#Sort Values
bblnrgdata_cut.sort_values(by = 'logUT', inplace = True)
```

```python
#Plot quadratic model over data
pl.figure(figsize=(10,10))
pl.plot(bblnrgdata_cut['logUT'],bblnrgdata_cut['lognrg'],'go')
pl.plot(bblnrgdata_cut['logUT'], OLS3.predict(bblnrgdata_cut),'-', color = 'orange')
pl.xlabel('Independent')
pl.ylabel('Depedent')
pl.title('Figure 11: Anscomby 1 - quadratic fit')
```

Figure nine shows energy vs total units per lot in log scale overlayed with a OLS quadratic fit. This was done by adding a new parameter of log(units total) squared. The fit appears to follow a trend of the data and looks to be the best fit seen thus far.

# Errors

```python
#Calculate errors
bblnrgdata_cut['errorsnrg'] = np.sqrt(bblnrgdata_cut['Reported_Property_Floor_Area_Buildings_ft2']**2 + bblnrgdata_cut['Site_EUI_kBTU_ft2']**2)
bblnrgdata_cut['errorsunits'] = np.sqrt(bblnrgdata_cut['UnitsTotal'])
```

```python
#Calculate log of erros
bblnrgdata_cut['errors_lognrg'] = bblnrgdata_cut['errorsnrg'] / (bblnrgdata_cut['nrg'] * np.log(10))
bblnrgdata_cut['errors_logunits'] = bblnrgdata_cut['errorsunits'] / (bblnrgdata_cut['UnitsTotal'] * np.log(10))
```

```python
#Plot data with errors
from matplotlib import gridspec

fig2 = plt.figure(figsize=(15,15))

gs1 = gridspec.GridSpec(3, 3)
gs1.update(left=0.68, right=0.68, wspace=0.2)

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=1)
ax1.errorbar(bblnrgdata_cut['lognrg'], bblnrgdata_cut['logUT'], yerr=bblnrgdata_cut['errors_logunits'], 
             fmt='.', alpha=0.5)
ax1.set_ylabel("log10 of Units", fontsize=15)
ax1.set_xlabel("log10 Energy", fontsize=15)


ax2 = plt.subplot2grid((3, 3), (0, 1), colspan=1)
ax2.errorbar(bblnrgdata_cut['logUT'], bblnrgdata_cut['lognrg'], yerr=bblnrgdata_cut['errors_lognrg'],
            fmt='.', alpha=0.5)
ax2.set_xlabel("log10 of Units", fontsize=15)
ax2.set_ylabel("log10 Energy", fontsize=15)

ax4 = plt.subplot2grid((3, 3), (1, 0), colspan=2, rowspan=2)
ax4.errorbar(bblnrgdata_cut['logUT'], bblnrgdata_cut['lognrg'], yerr=bblnrgdata_cut['errors_lognrg'],
           xerr=bblnrgdata_cut['errors_logunits'], alpha=0.5, fmt='.')
ax4.set_xlabel("log10 of Units", fontsize=15)
ax4.set_ylabel("log10 Energy", fontsize=15)
ax4.set_ylim(5,11)

print ('The largest error bar is for')
bblnrgdata_cut[bblnrgdata_cut['errorsnrg'] == bblnrgdata_cut['errorsnrg'].max()]
```

# Add the Errors and Calculate the Chi Square

```python
#Define Chi Sqaured Function

'''
This function is to calculate the chi squared value via the residuals and errors associated with the data. 
First the residuals are calculated and then divided by the errors. Finally, they are summed to find the final chi
sqaured value. 

The strength of the chi squared shows how well the model is fitting the data. Lower residuals means a good fit, 
where higher residuals, especially those associated with error, may produce larger chi squared values due to higher
variances.

'''
def chi2(data, model, errors = None):
    if errors is not None:
        chisq = (((data - model) ** 2) / errors ** 2).sum().sum()
        return chisq
    else:
        chisq = ((data - model) ** 2 ).sum()
        return chisq
```

```python
#Total Error = sqrt(error1^2 + error2^2)
bblnrgdata_cut['total_error'] = sqrt((bblnrgdata_cut['errors_lognrg']*bblnrgdata_cut['errors_lognrg'])+
                                     (bblnrgdata_cut['errors_logunits']*bblnrgdata_cut['errors_logunits']))
```

```python
#degrees of freedom: number of observations - number of model parameters
# note: these are reported as dof of the residuals in the statsmodels fit summary, but jsut to make sure let's calculate them
dof = len(bblnrgdata_cut)-len(OLS1.params)
print ("degrees of freedom:" , dof)
print(OLS2.df_resid)
```

```python
print ("Units vs Energy residuals (no errors include): %.2f"%\
       (chi2(bblnrgdata_cut['lognrg'], OLS1.predict()) / 
        float(dof)))

print ("Units vs Energy chi square per DOF w independent variable error only: %.2f"%\
       (chi2(bblnrgdata_cut['lognrg'], OLS1.predict(), 
            errors = bblnrgdata_cut['errors_lognrg']) / dof))

print ("Units vs Energy chi square per DOF : %.2f"%\
       (chi2(bblnrgdata_cut['lognrg'], OLS1.predict(), 
            errors = np.sqrt(bblnrgdata_cut['errors_lognrg']**2 + 
                             bblnrgdata_cut['errors_logunits']**2))  / dof))
```

```python
print ("Units vs Energy residuals (no errors include): %.2f"%\
       (chi2(bblnrgdata_cut['logUT'], OLS2.predict()) / 
        float(dof)))

print ("Units vs Energy chi square per DOF w independent variable error only: %.2f"%\
       (chi2(bblnrgdata_cut['logUT'], OLS2.predict(), 
            errors = bblnrgdata_cut['errors_lognrg']) / dof))

print ("Units vs Energy chi square per DOF : %.2f"%\
       (chi2(bblnrgdata_cut['logUT'], OLS2.predict(), 
            errors = np.sqrt(bblnrgdata_cut['errors_lognrg']**2 + 
                             bblnrgdata_cut['errors_logunits']**2))  / dof))
```

OLS 1 is the linear model that appeared to fit the data better compared to OLS2. The chi squared values tell the same thing. Per each test, with no errors, errorrs, and degrees of freedom, the OLS1 has associated chi squared values that are lower than the ones produced from OLS2, showing that OLS1 is a better fit.

Figure 13 shows the two influence plots for OLS1 and OLS2. They show the influence the residuals of all points have on the OLS models. The larger residuals, the more influence. For OLS1 the largest infleucne appears to be at -3 studentized residuals, while -6 for OLS2. Since OLS1 is a much stronger models, these figures aid in stating that, where OLS1 shows smaller residuals,despite larger influence amongst the data points.
