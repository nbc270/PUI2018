```python
january_data = pd.read_csv(os.getenv("PUIDATA") + '/January2017_citibike_tripdata.csv.zip')
august_data = pd.read_csv(os.getenv("PUIDATA") + '/August2017_citibike_tripdata.csv.zip')
```

```python
#january_data is the dataframe where the content of the January csv file is stored
january_data.head()
```

```python
#August_data columns needed to match January's columns
august_data.columns = ([u'Trip Duration', u'Start Time', u'Stop Time', u'Start Station ID',
       u'Start Station Name', u'Start Station Latitude',
       u'Start Station Longitude', u'End Station ID', u'End Station Name',
       u'End Station Latitude', u'End Station Longitude', u'Bike ID',
       u'User Type', u'Birth Year', u'Gender'])

#august_data is the dataframe where the content of the August csv file is stored
january_data.head()
```

```python
#Concatenating the two dataframes
combined_data = pd.concat([january_data, august_data], sort=True)
combined_data.head()
```

```python
#See how much data there is before removing NaN
print(len(combined_data))
```

```python
#Removing NaN and see how much data remains
combined_data.dropna(inplace=True)
print(len(combined_data))
```

```python
#Adding a new date column, where I can extract the time
combined_data['date'] = combined_data[u'Start Time'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
combined_data.head()
```

```python
# Get the hour of the trip starttime
combined_data[u'Start Hour'] = combined_data.date.apply(lambda x: x.hour)
combined_data.columns
```

```python
# I am defining day hours as 6am - 5pm (6-17) and night hours as 6pm - 5am (18-5)

# Add a time_of_day column to identify day vs. night trips using mapping
start_hour = {6: 'day', 7: 'day',8: 'day', 9: 'day',10: 'day',11: 'day',12: 'day',13: 'day',14: 'day', 15: 'day',
                          16: 'day',17: 'day',18: 'night', 19: 'night',20: 'night',21: 'night', 22: 'night', 23: 'night',
                          0: 'night',1: 'night', 2: 'night', 3: 'night', 4: 'night',5: 'night'}

combined_data['Hour_Began'] = combined_data[u'Start Hour'].map(start_hour)
combined_data['Hour_Began'] = combined_data[u'Start Hour'].map(start_hour)
combined_data.head()
```

```python
#Create columns for the ride rudation for day time and night time. Fill opposing data with NaN

combined_data['Day_Ride'] = np.where(combined_data['Hour_Began'] == 'day', combined_data['Trip Duration'], np.NaN)
combined_data['Night_Ride'] = np.where(combined_data['Hour_Began'] == 'night', combined_data['Trip Duration'], np.NaN)
combined_data.head(200)
```

```python
combined_data.columns
```

```python
#Reduced data
combined_data = combined_data[[u'Day_Ride',u'Night_Ride']]
combined_data.tail()
```

```python
combined_data[u'Day_Ride'].dropna(inplace= True)
combined_data[u'Night_Ride'].dropna(inplace= True)
```

```python
# Sub jectively choosing an interval of 50 minutes with bins at every 3 minutes (180 seconds)
bins = np.arange(0,3000, 180)
day = combined_data.Day_Ride.groupby(pd.cut(combined_data.Day_Ride, bins)).agg([count_nonzero]).plot(kind='bar', 
                                                                legend=False)
day.set_title("Day RIde Duration")
day.set_xlabel("Duration of Rides (seconds)")
day.set_ylabel("Number of Rides")

night = combined_data.Night_Ride.groupby(pd.cut(combined_data.Night_Ride, bins)).agg([count_nonzero]).plot(kind='bar', 
                                                                legend=False)
night.set_title("female riders")
night.set_xlabel("Duration of Rides (seconds)")
night.set_ylabel("Number of Rides")
```

Figures 1&2 are histograms of day (top) and night (bottom) rides amongst citibike riders. We see that for both riders, rides tend to las t between 180 and 540 seconds (3-9 minutes).

```python
#Group the data by day and night to nromalize it and begin platting it over every three miniutes.

csD=combined_data.Day_Ride.groupby(pd.cut(combined_data.Day_Ride, bins)).agg([count_nonzero]).cumsum()

csN=combined_data.Night_Ride.groupby(pd.cut(combined_data.Night_Ride, bins)).agg([count_nonzero]).cumsum()

print (np.abs(csD / csD.max()-csN / csN.max()))

pl.plot(bins[:-1] + 5, csD / csD.max(), label = "Day")
pl.plot(bins[:-1] + 5, csN / csN.max(), label = "Night")
pl.plot(bins[:-1] + 5, np.sqrt(csN / csN.max() - csD / csD.max())**2, 'k-',
        label = "difference")
pl.xlabel("Rides")
pl.ylabel("Normalized Cumulative Number")
pl.title("Figure 3: Normalized Cummulative Number for Both Day and Night Rides ")
pl.legend()
```

Figure 3 show the nromalized count curves of rides for both day and night as well as the difference between the two. As the number of rides grows, the nromalized data begins to plateau. The curves are very similar sguggesting a relationship or at least similar behavior between day and night riders.

# KS Test
# H0 = Samples derive from the sample distribution
# H1 = Samples derive from different distributions

```python
#KS-Test for whole samples
ks = stats.ks_2samp(combined_data.Day_Ride, combined_data.Night_Ride)
print(ks)
```

Since the ks-statistic reported by the KS-Test is almost zero, we fail to reject the null hypothesis. Even though there is a low p-value that would suggest rejecting the null, we cannot due to the low ks-statistic.

```python
# Creating a smaller sample
np.random.seed(123)

#Convert to numpy arrays
day_array = np.array(combined_data['Day_Ride'])
night_array = np.array(combined_data['Night_Ride'])
print(len(day_array))
print(len(night_array))
```

```python
#Make the arrays 200th of the size
random.seed(1229)
new_day = np.sort(np.random.choice(day_array, size = len(day_array)/200, replace = False))
new_night = np.sort(np.random.choice(night_array, size = len(night_array)/200, replace = False))

day_length = len(new_day)
night_length = len(new_night)
print(day_length)
print(night_length)
```

```python
#Perform KS Test on smaller sample
ks = stats.ks_2samp(new_day, new_night)
print(ks)
```

Now that we have a reduced data set, we see the ks-statistic reported by the KS-Test is almost zero & the p-value is greater than 0.05. Therefore, we fail to reject the null hypothesis.

# Pearsons Test

# H0: There is no correlation between day trips and night trips 

```python
#alpha = 0.05
```

```python
#Arrays have to be same size. Randomly choose 600,000 ride values for day and night
day_array_sized = random.choice(new_day,3500)
night_array_sized = random.choice(new_night,3500)
```

```python
# Sort the data (low to high)
np.sort(day_array_sized)
np.sort(night_array_sized)
```

```python
#Perform Pearsons Test on the sorted and randomly selected arrays of the same size
stats.pearsonr(day_array_sized, night_array_sized)
```

Based on the correlation coeefieicnt's absolute value, 0.008, the correlation coefficient is considered 'very wea'. The p-value is also much larger than 0.05 so we fail to reject the null hypothesis.

# Spearman's Test

# H0: There is no monotonic (function never increases or decreases with change of independent variable) relationship between day trip and night trip durations

Spearmans Hypothesis Testing Info: http://www.statstutor.ac.uk/resources/uploaded/spearmans.pdf

```python
#alpha = 0.05
```

```python
#Perform spearmans Test on the sorted and randomly selected arrays of the same size
stats.spearmanr(day_array_sized,night_array_sized)
```

Since the absolute value of the correlation coeffiecient is 0.011, the correlation coefficient is considered 'very weak'. The p-value is already high enough that we cannot accept the null hypothesis that day trip and night trips are not monotonic.
